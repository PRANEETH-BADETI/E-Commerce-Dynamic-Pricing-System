```python
columns_to_clean = [
    'title', 'condition_0_conditionDisplayName', 'isMultiVariationListing',
    'listingInfo_0_watchCount', 'primaryCategory_0_categoryName',
    'returnsAccepted', 'shippingInfo_0_handlingTime', 'shippingInfo_0_oneDayShippingAvailable',
    'topRatedListing'
]

for col in columns_to_clean:
  df_new[col] = df_new[col].str.strip('[]').str.replace("'","")
```

```python
numeric_cols = ['listingInfo_0_watchCount','shippingInfo_0_handlingTime']

for col in numeric_cols:
  df_new[col] = pd.to_numeric(df_new[col])
```

```python
bool_columns = ['isMultiVariationListing', 'returnsAccepted', 'shippingInfo_0_oneDayShippingAvailable', 'topRatedListing']
for col in bool_columns:
    df_new[col] = df_new[col].map({'true': 1, 'false': 0})
```

```python
df_new = pd.get_dummies(df_new,columns = category_cols, drop_first = True)
```

```python
# Create a normalized discount feature
df_new['Normalized_Discount'] = df_new['Discount_amount'] / (df_new['discountPriceInfo_0_originalRetailPrice_0___value__'] + 1e-9)

# Create a demand score
df_new['Demand_Score'] = df_new['listingInfo_0_watchCount'] / (df_new['shippingInfo_0_handlingTime'] + 1e-9)
```

```python
dataset = pd.read_csv('/content/Clean_data.csv')
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,accuracy_score
```

```python
#Final Data Preprocessing
```

```python
columns_to_drop = ['Unnamed: 0', 'title']
dataset = dataset.drop(columns = columns_to_drop)
```

```python
# Handle outliers
# Cap Demand_Score and listingInfo_0_watchCount to the 99th percentile
cap_threshols = ['Demand_Score', 'listingInfo_0_watchCount']
for col in cap_threshols:
  upperlimit = dataset[col].quantile(0.99)
  dataset[col] = np.where(dataset[col]>upperlimit, upperlimit,dataset[col])
```

```python
# Normalize numeric columns (using MinMaxScaler)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numeric_cols = dataset.select_dtypes(include=['int64', 'float64']).columns
dataset[numeric_cols] = scaler.fit_transform(dataset[numeric_cols])
```

```python
# Display final cleaned dataset
print("Data after preparation:")
print(dataset.head())
print(f"Shape of the data: {dataset.shape}")
```

```python
target = 'sellingStatus_0_convertedCurrentPrice_0___value__'
x = dataset.drop(columns = [target])
y = dataset[target]
```

```python
# Split dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor(n_estimators = 100, random_state = 42)
model.fit(X_train,y_train)
```

```python
# 6ï¸âƒ£ **Make Predictions**
y_pred = model.predict(X_test)

# 7ï¸âƒ£ **Evaluate the Model**
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)

print(f"ðŸ“ˆ **Model Performance:**")
print(f"MAE (Mean Absolute Error): {mae:.2f}")
print(f"MSE (Mean Squared Error): {mse:.2f}")
print(f"R2 Score (Coefficient of Determination): {r2:.2f}")
```
